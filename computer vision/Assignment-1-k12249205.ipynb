<li>Use the function <i><b>create_Gaussian</b><i> to create the Gaussian filter in the frequency domain.</li>
<li>The shape of the <i><b>created Gaussian filter</b><i> should be the same as the shape of the grayscale image <i>zebras<i>.</li>
</ul>

```python
#%matplotlib notebook
from io import BytesIO
from IPython.display import clear_output, Image as NoteImage, display
def imshow(im,fmt='jpeg'):
    #a = np.uint8(np.clip(im, 0, 255))
    f = BytesIO()
    PIL.Image.fromarray(im).save(f, fmt)
    display(NoteImage(data=f.getvalue()))
```

```python
def create_Gaussian(h :int, w: int, cutoff : int ):
    H = np.zeros((h,w), dtype=np.float32)
    for u in range(h):
        for v in range(w):
            D = np.sqrt((u-h/2)**2 + (v-w/2)**2)
            H[u,v] = np.exp(-D**2/(2*cutoff*cutoff))
    return H
```

```python
# The code goes here
from scipy.fft import fft2, fftshift, ifft2


def create_Gaussian(shape, cutoff):
    row, col = shape
    
    x = np.linspace(-col / 2, col / 2, col)
    y = np.linspace(-row / 2, row / 2, row)
    X, Y = np.meshgrid(x, y)
    gaussian_filter = np.exp(-((X**2 + Y**2) / (2 * (cutoff**2))))
    return gaussian_filter

gaussian_filter = create_Gaussian(zebra_gray.shape, 40)
zebra_gray_freq = fft2(zebra_gray)
zebra_gray_freq_shft = fftshift(zebra_gray_freq)
```
